```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(x,y)
axes2.plot(y,x,'r')
axes2.set_title('Smaller')
axes1.set_title('Larger')
```

```python
fig,axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
axes
```

```python
fig = plt.figure(figsize = (4,2), dpi =100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize = (4,2), nrows=2, ncols=1)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig.savefig('my_picture.png')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'x^2')
ax.plot(x,x**3, label = 'x^3')
ax.plot(x,x**4, label = 'x^4')

ax.legend(loc = 0)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'blue', lw=3,
        ls = '--', marker = 'o', markersize=12,
       markerfacecolor='red', markeredgewidth=3,
       markeredgecolor='green')
```

```python
from math import sin
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,sin(x))
```
